#                    **Introduction**

The main purpose of this project is to build an AI system that can automatically detect potential greenwashing in companies’ ESG-related text. Greenwashing happens when organizations make their environmental or sustainability performance sound better than it truly is, for example by using vague promises, marketing language, or selective reporting instead of concrete actions and measurable results. As ESG reports, sustainability statements, and corporate disclosures become more common, it is increasingly difficult for investors, regulators, and the public to manually evaluate which claims are genuine and which are misleading.

In this project, we focused on the specific task of classifying ESG statements as either genuine or misleading based on their text. The dataset consists of around 860 rows of preprocessed ESG content, along with additional numeric features such as environmental, social, and governance scores.

To model this task, we fine-tune FinBERT, a **transformer model pretrained** on financial and business text, as a binary classifier on my labeled dataset. FinBERT is well suited to this domain because ESG and sustainability disclosures are written in a similar style to financial reports and corporate communications. The final goal is to train a model that takes an ESG statement as input and outputs a prediction of whether the claim is more likely to be genuine or misleading, along with standard evaluation metrics.

Overall, this project we demonstrated how modern NLP techniques and domain-specific language models can be used to support automatic greenwashing detection, providing a first-pass tool to flag risky or suspicious sustainability claims for further human review.

# **Hypothesis-1**
There is a significant correlation between Environmental (E), Social (S), and Governance (G) scores.

**Goal:**
To check if companies that perform well in one ESG dimension (e.g., Environmental) also perform well in others (Social, Governance).

**Statistical Method:**

**Pearson Correlation Test** — measures the linear relationship between two continuous variables to verify.

**Null Hypothesis (H₀)**: There is no significant correlation between E, S, and G scores.

**Alternative Hypothesis (H₁)**: There is a significant correlation between at least two of the ESG components.

```python
from scipy.stats import pearsonr

# Select relevant columns
esg_scores = df_esg[['e_score', 's_score', 'g_score']].dropna()

# Compute correlation matrix
corr_matrix = esg_scores.corr(method='pearson')
print("Correlation Matrix:\n", corr_matrix)

# Run pairwise Pearson correlation tests
for col1 in esg_scores.columns:
    for col2 in esg_scores.columns:
        if col1 != col2:
            corr, p_value = pearsonr(esg_scores[col1], esg_scores[col2])
            print(f"\nCorrelation between {col1} and {col2}:")
            print(f"r = {corr:.3f}, p-value = {p_value:.4f}")

# Visualization - Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation between E, S, and G Scores")
plt.show()
```

**Findings form H1**

Pearson tests show no significant correlation between E and S (r≈−0.03, p≈0.33), a weak negative correlation between E and G (r≈−0.16, p<0.001), and a moderate positive correlation between S and G (r≈+0.41, p<0.001). This indicates S and G tend to move together, while E behaves somewhat independently (and even slightly inversely with G).

# **Hypothesis-2**

Average overall ESG scores have increased over the years (2020–2022), showing an improvement in sustainability reporting.

**Goal:**
To test whether the mean total ESG scores differ significantly across different years.

**Statistical Method:**

**One-Way ANOVA (Analysis of Variance)** — compares the mean of a continuous variable across three or more groups (in this case, years).

**Null Hypothesis (H₀)**: There is no significant difference in average ESG total scores between years.

**Alternative Hypothesis (H₁)**: There is a significant difference in average ESG total scores between years (at least one year differs).

```python
from scipy.stats import f_oneway

# Clean and convert the 'year' column
df_esg['year'] = df_esg['year'].astype(str).str.extract('(\d{4})')
df_esg = df_esg.dropna(subset=['year', 'total_score'])
df_esg['year'] = df_esg['year'].astype(int)

# Group data by year
grouped = df_esg.groupby('year')['total_score']

# Perform ANOVA test
anova_result = f_oneway(*(grouped.get_group(year) for year in grouped.groups))
print("ANOVA Test Result:")
print(f"F-statistic = {anova_result.statistic:.3f}, p-value = {anova_result.pvalue:.4f}")

# Visualization - Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x='year', y='total_score', data=df_esg, palette='Set2')
plt.title('Distribution of Total ESG Scores by Year')
plt.xlabel('Year')
plt.ylabel('Total ESG Score')
plt.show()

df_esg.groupby('year')['total_score'].mean().plot(marker='o', color='green', figsize=(6,4))
plt.title("Average ESG Total Score Over Years")
plt.xlabel("Year")
plt.ylabel("Mean Total ESG Score")
plt.show()
```

**Findings from H2**

One-way ANOVA shows significant differences in mean total ESG scores across years (F=3.118, p=0.001). The mean-by-year line plot indicates an overall upward trend, consistent with improving ESG reporting over time.

# **Hypothesis-3**

**Statement:**
Companies with higher Environmental (E) scores also tend to have higher overall ESG (Total) scores.

**Goal:**
To test whether the Environmental score (e_score) has a significant linear relationship with the Total ESG score (total_score) — i.e., does strong environmental performance predict overall ESG strength?

**Statistical Method:**

**Simple Linear Regression**- examines the relationship between one independent variable (E score) and one dependent variable (Total ESG score).

**Null Hypothesis (H₀):** There is no linear relationship between Environmental score and Total ESG score.

**Alternative Hypothesis (H₁):** There is a significant positive linear relationship between Environmental score and Total ESG score.

```python
from scipy.stats import linregress

df_clean = df_esg.dropna(subset=['e_score','total_score'])
res = linregress(df_clean['e_score'], df_clean['total_score'])
print(f"Slope = {res.slope:.3f}")
print(f"Intercept = {res.intercept:.3f}")
print(f"r = {res.rvalue:.3f}, R^2 = {res.rvalue**2:.3f}, p = {res.pvalue:.4f}")

plt.figure(figsize=(6,4))
sns.regplot(x='e_score', y='total_score', data=df_clean, scatter_kws={'alpha':0.5})
plt.title('E score vs Total ESG score')
plt.xlabel('Environmental (E) score')
plt.ylabel('Total ESG score')
plt.show()
```

**Findings from H3:**

The Simple Linear Regression shows a significant positive relationship between the Environmental (E) score and the Total ESG score, with a p-value < 0.05.
This means that companies with higher Environmental performance generally achieve higher overall ESG ratings.
The positive slope and moderate R² value confirm that environmental efforts contribute meaningfully to total ESG outcomes, supporting the alternative hypothesis (H₁).

Converting ner_entities from String to List
The ner_entities column contains stringified Python lists — each entry looks like a list of named entities but is stored as a string.


```python
import ast
import pandas as pd

def safe_parse(x):
    """
    Safely convert string to list.
    Returns [] if conversion fails.
    """
    # Handle NaN or None
    if pd.isna(x):
        return []

    # Convert to string
    x = str(x)

    # Try literal_eval
    try:
        return ast.literal_eval(x)
    except:
        # If fails, return empty list
        return []

# Apply the safe parser to the column
df_esg['ner_entities'] = df_esg['ner_entities'].apply(safe_parse)

print("✔ Fixed: 'ner_entities' column successfully converted into lists.")
```

```python
ESG_KEYWORDS = {
    "environmental": [
        "climate change", "carbon footprint", "greenhouse gas", "ghg emissions", "net zero",
        "emissions reduction", "renewable energy", "solar power", "wind energy", "hydropower",
        "sustainable packaging", "waste management", "energy efficiency", "biodiversity",
        "recycling program", "sustainable sourcing", "scope 1 emissions", "scope 2 emissions",
        "scope 3 emissions", "deforestation", "pollution control", "air quality", "water conservation",
        "circular economy", "eco-friendly", "carbon neutrality", "environmental impact", "environmental stewardship",
        "clean energy", "resource conservation", "sustainable agriculture", "battery recycling"
    ],
    "social": [
        "diversity and inclusion", "gender equality", "equal opportunity", "human rights",
        "labor standards", "fair wages", "living wage", "employee wellbeing", "occupational safety",
        "workplace safety", "mental health support", "employee engagement", "training and development",
        "volunteer programs", "community engagement", "social impact", "supply chain transparency",
        "non-discrimination", "health and safety", "stakeholder engagement", "employee rights",
        "local communities", "employee retention", "child labor", "forced labor", "workers’ rights",
        "education access", "workplace culture", "employee satisfaction", "community investment"
    ],
    "governance": [
        "corporate governance", "board diversity", "executive compensation", "audit committee",
        "independent directors", "shareholder rights", "business ethics", "anti-corruption",
        "compliance program", "internal audit", "transparency", "risk management", "whistleblower protection",
        "conflict of interest", "data privacy", "cybersecurity governance", "regulatory compliance",
        "board independence", "sustainability oversight", "esg committee", "stakeholder governance",
        "bribery prevention", "code of conduct", "ethics training", "governance structure", "voting rights"
    ]
}
```

Keyword Counter (Case-Insensitive, Whole Phrases)
We define a function to count keyword occurrences for each ESG pillar in a given report.

Matching is done in a case-insensitive manner.
We use whole-phrase matching with word boundaries to avoid false positives (e.g., "net" in "internet").
This function will be applied to each report in the dataset to extract interpretable content-level features.

```python
# Function to count ESG keyword matches in a single text
def count_esg_terms(text: str, keyword_dict: dict) -> dict:
    counts = defaultdict(int)
    text_lower = text.lower()

    for pillar, terms in keyword_dict.items():
        for term in terms:
            # Use word boundaries to match full phrases only
            pattern = r"\b" + re.escape(term.lower()) + r"\b"
            matches = re.findall(pattern, text_lower)
            counts[pillar] += len(matches)

    return counts

# Confirmation print
print(" ESG keyword counting function defined.")
```

Compute ESG Keyword Frequencies for Each Report
We now apply the keyword counting function to every report in the dataset, producing three new features:

env_kw: number of environmental keyword mentions
soc_kw: number of social keyword mentions
gov_kw: number of governance keyword mentions
These will serve as interpretable, pillar-specific content features for comparison with ESG scores.

```python
# Apply ESG keyword counter to each report and store results in new columns
df_esg[["env_kw", "soc_kw", "gov_kw"]] = (
    df_esg["preprocessed_content"]
    .apply(lambda txt: count_esg_terms(txt, ESG_KEYWORDS))
    .apply(pd.Series)
    .fillna(0)
    .astype(int)
)

print("ESG keyword frequencies computed for all reports.")
```

Normalize Keyword Counts by Report Length (per 1,000 Words)
To ensure fair comparisons across reports of varying lengths, we normalize the raw keyword counts by total word count.

This gives us ESG keyword mention rates per 1,000 words, allowing us to distinguish between genuinely ESG-focused reports and those that simply happen to be long.

```python
# Normalize keyword counts by report length (per 1,000 words)
df_esg["words_per_1k"] = df_esg["word_count"] / 1000.0

for col in ["env_kw", "soc_kw", "gov_kw"]:
    # Create new columns like env_kw_rate, soc_kw_rate, gov_kw_rate
    df_esg[f"{col}_rate"] = df_esg[col] / df_esg["words_per_1k"]

print("ESG keyword rates normalized per 1,000 words.")
```

# **Deep Learning Model for Greenwashing Detection using finbert**


To model this task, I fine-tune FinBERT, a transformer model pretrained on financial and business text, as a binary classifier on my labeled dataset. FinBERT is well suited to this problem because ESG and sustainability disclosures are written in a style that is very similar to financial reports and corporate communications. The final goal is to train a model that takes an ESG statement as input and predicts whether the claim is more likely to be genuine or misleading, and then evaluate its performance using standard classification metrics.

```python
!pip install -q "transformers>=4.30" datasets evaluate

import numpy as np
import pandas as pd
import evaluate

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)
```

```python
file_name = "DATA602_ESG_Dataaset"
text_col  = "preprocessed_content"
label_col = "label"
df_esg = pd.read_csv(file_name)

label_map = {"genuine": 0, "misleading": 1}
df_esg["label_id"] = df_esg[label_col].map(label_map)

train_df, temp_df = train_test_split(
    df_esg,
    test_size=0.3,
    stratify=df_esg["label_id"],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["label_id"],
    random_state=42
)
```

```python
MODEL_NAME = "yiyanghkust/finbert-tone"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def tokenize_function(examples):
    return tokenizer(
        examples[text_col],
        padding="max_length",
        truncation=True,
        max_length=256
    )

train_dataset = Dataset.from_pandas(train_df[[text_col, "label_id"]])
val_dataset   = Dataset.from_pandas(val_df[[text_col, "label_id"]])
test_dataset  = Dataset.from_pandas(test_df[[text_col, "label_id"]])

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_val   = val_dataset.map(tokenize_function, batched=True)
tokenized_test  = test_dataset.map(tokenize_function, batched=True)

tokenized_train = tokenized_train.rename_column("label_id", "labels")
tokenized_val   = tokenized_val.rename_column("label_id", "labels")
tokenized_test  = tokenized_test.rename_column("label_id", "labels")

tokenized_train.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_val.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_test.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

In this step, I prepare the text data so it can be used by FinBERT. I load the FinBERT tokenizer, convert the train/validation/test splits into Hugging Face datasets, tokenize the preprocessed_content text, rename label_id to labels, and set the format to PyTorch tensors (input_ids, attention_mask, labels) for training.

```python
# 1. Model (FinBERT with 2 classes)
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=2,
    ignore_mismatched_sizes=True
)

# 2. Metric
metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

# 3. Training settings (no wandb)
training_args = TrainingArguments(
    output_dir="./finbert_greenwashing_model",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_steps=50,
    seed=42,
    report_to="none",

# 4. Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    compute_metrics=compute_metrics,
)

print("Starting Training...")
trainer.train()
print("Training Complete!")
```

In this step, I fine-tune FinBERT as a binary classifier by loading the pre-trained model with two output classes and defining accuracy as the evaluation metric. I then specify the main training hyperparameters (epochs, batch size, learning rate) in TrainingArguments, create a Trainer with the tokenized datasets and metrics function, and run trainer.train() to adapt FinBERT to the greenwashing detection task.

```python
preds = trainer.predict(tokenized_test)
logits = preds.predictions
y_pred = np.argmax(logits, axis=-1)
y_true = np.array(test_df["label_id"])

print(classification_report(y_true, y_pred, target_names=["genuine","misleading"], digits=4))
print(confusion_matrix(y_true, y_pred))
```
